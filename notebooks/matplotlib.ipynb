##### A simple plot

```python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='My first plot')
#ax.grid()

print(fig.canvas.get_supported_filetypes())

fig.savefig("test.png")
plt.show()
```

```python
fig.savefig('test80.png', transparent=False, dpi=80, bbox_inches="tight")
fig.savefig('test300.png', transparent=True, dpi=300, bbox_inches="tight")
```

##### Histograms

```python
np.random.seed(10808691)
# example data
mu = 10  # mean of distribution
sigma = 0.5  # standard deviation of distribution
x = mu + sigma * np.random.randn(867)
```

```python
fig, ax = plt.subplots()

# the histogram of the data
num_bins = 50
n, bins, patches = ax.hist(x, num_bins, density=1)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, 'o')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=10$, $\sigma=0.5$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
